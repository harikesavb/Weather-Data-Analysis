```python
import pandas as pd
#Reading the csv file
df=pd.read_csv('/content/Weather Data_set.csv')
```

```python
print(df.shape)
```

```python
df.head()
```

```python
df.tail()
```

```python
#check For null values and drop them
print(df.isnull().sum())
df.dropna() #drop any row with NaN
df.dropna(axis=1) #drop any column with NaN
```

```python
#unique weather types
weather=df['Weather'].value_counts()
dfweather=pd.DataFrame(weather)
dfweather = dfweather.reset_index()
dfweather.columns = ['Weather', 'Frequency'] # change column names
dfweather
```

```python
#rename column named 'Weather' to Weather Condition'
df = df.rename(columns = {'Weather' : 'Weather_Condition'}, inplace=False)
```

```python
df['Weather_Condition']
```

```python
#all records from weather was exactly clear
df[df['Weather_Condition'] == 'Clear']
```

```python
#mean temperature, wind speed and visibility
print(df['Temp_C'].mean())
print(df['Wind Speed_km/h'].mean())
print(df['Visibility_km'].mean())
```

```python
#variance of pressure
print(df['Press_kPa'].var())
```

```python
#the days on which wind speed was less than or equal to 30 km/hr and temperature was greater than 0 C
windspeed=df[(df['Wind Speed_km/h'] <=30) & (df['Temp_C']>0)]
print(windspeed['Date/Time'])
```

```python
#the date and temperatures for all instances when snow was recorded
snow=df[df['Weather_Condition'] == 'Snow']
snow.loc[:, ['Date/Time', 'Temp_C']]
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#display a graph of variation of temperature with respect to time
graph=df[['Temp_C', 'Date/Time']]
graph.plot(x='Date/Time', y='Temp_C',figsize=(10,7) )
```

```python
#display a pie chart of percentages of weather conditions
keep=dfweather[dfweather['Frequency']>20]
print(keep)
keep.plot.pie(y='Frequency',autopct='%1.1f%%', shadow=False, figsize=(12,9))
plt.title("Weather Conditions")
plt.show()
```
